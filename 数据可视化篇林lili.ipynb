### 统计学基础：

数据的分类：分类数据，数值数据

#### 分类数据：表示有限数量类别的数据(如：性别)

分为定序数据和定类数据

定序数据表示的数据是有顺序的(如：金银铜牌可以按顺序排列)

定类数据没有顺序

#### 数值数据：表示数值的数据，可以进行数学运算和统计分析(如:身高体重值)

数值数据分为连续数据和离散数据

连续数据中任意两个数据之间可以取无限多的值(如：身高体重)

离散数据标识的数据只能以整数或自然数为单位(如：人口数量)

### 数值数据分析维度：

##### 集中趋势

数据居中在何处？

三个指标：平均数，中位数，众数

平均数：`df1['列名'].mean()`

中位数：`df1['列名'].median()`

众数：`df1['列名'].mode()`
##### 离散趋势

数据偏离中心的散布情况

三个指标：极差，方差(标准差)，四分位距

极差=最大值-最小值

`df1['列名'].max()=df1['列名'].min()`

`np.array(df1['列名']).ptp()`#ptp方法直接求极差

方差：`df1['列名'].var()`

标准差：`df1['列名'].std()`

四分位距：第三四分位数-第一四分位数

`q1=df1['列名'].quantile(0.25)`

`q3=df1['列名'].quantile(0.75)`

四分位距：`q3-q1`

##### 分布形状

数据的分布对称程度，峰值高低等等

绘制数据的分布直方图：`df1['列名'].plot(kind='hist')`

#### 常见图表：

直方图(一维数据里属于各个数据范围内有多少个数值)

密度图(表示数据的分布密度)

箱型图(表示数据的四分位数分布情况)

    上界=min(第三四分位数+1.5*四分位距,最大值)(从两者中取小值)
    下界同理
    若最大(小)值大于第三(yi)四分位数+(-)1.5*四分位距，则说明最大(小)值为异常值，用单独的点表示·
    
小提琴图

小提琴两大侧面积表示为数值密度(参考密度图)

中间小圆点表示中位数值

中间箱子表示四分位数范围(参考箱线图)

中间黑线表示在95％水平置信区间内数值
