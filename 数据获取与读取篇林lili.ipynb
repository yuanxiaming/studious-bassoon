##### json数据格式：

json布尔值均为小写字母：true，false

json对象类似于python字典，json数组类似于python列表

json对象：以大括号包围，内部为键和值，每组键值之间通过逗号隔开，

键必须独立且唯一，必须是字符串，必须用双引号包围


{

"键":值，

"键":值,

"键":值,

}

json数组：

[值1，值2，值3]

##### pandas读取json文件：
```python
import pandas as pd

pd.read_json("文件路径")

```

##### pandas读取csv格式文件：
```

pd.read_csv("文件路径",header= ，index_col="指定为标签索引的列名")

```

header=None 表示不要把第一行当作表头

虽然read_csv默认会把第一行当作表头(列名)，但不会把第一列当作索引，而是用默认的位置索引

index_col可以手动设置指定的一列作为标签索引，在双引号内部填入想要指定索引标签的列名

##### 设置展示设置：pd.set_option()

`pd.set_option("display.max_columns",150)`  #设置展示列数的上限为150

`pd.set_option("display.max_colwidth",550)`#设置展示单元格字符长度的上限为550

df1.head(5)设置只读取前5行数据
